```python
ratios = [pos / neg for pos, neg in zip(result, negresults)]


# 创建图表
plt.figure(figsize=(10, 5))

# 绘制点线图
plt.plot(years, ratios, marker='o', linestyle='-', color='blue', label='Positive/Negative Ratio')

# 添加数据点数值
for (x, y) in zip(years, ratios):
    plt.text(x, y, f'{y:.2f}', color='red', fontsize=10, ha='center', va='bottom')

# 添加图例
plt.legend()

# 添加标题和标签
plt.title('Ratio of Positive to Negative Ratings by Year')
plt.xlabel('Year')
plt.ylabel('Positive/Negative Ratio')

# 显示网格
plt.grid(True)

# 显示图表
plt.show()
```
